```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import torch
```

```python
df = pd.read_csv('Datasets/NASA_turbofan_get_engine/train_FD001.txt', sep=' ', header= None)
# df.columns = ['unit number', 'time/cycle', 'op set 1', 'op set 2', 'op set 3']
df = df.drop(columns=[26,27])
df.dropna(inplace=True)
df.head()
```

```python
machines = df[0].unique()
```

```python
dic = {}
for i in machines:
    dic[i] = df[df[0] == i].shape[0]
```

```python
def filter_above_average(dictionary):
    values = list(dictionary.values())
    average = sum(values) / len(values)
    above_average_values = {key: value for key, value in dictionary.items() if value > average}
    return above_average_values
result = filter_above_average(dic)
len(result)
```

```python
data = []
label = []
for machine in result:
    filtered = df[df[0]==machine]
    filtered.drop(columns= [0, 1], inplace=True)
    np_array = filtered.to_numpy()
    tensor = torch.tensor(np_array[-100:], dtype=torch.float).unsqueeze(0)
    data.append(tensor)
    # appending 1 for failing
    label.append(torch.tensor([1]))
    tensor = torch.tensor(np_array[-200:-100], dtype=torch.float).unsqueeze(0)
    data.append(tensor)
    label.append(torch.tensor([0]))

final_tensor = torch.cat(data, dim=0)
label_tensor = torch.cat(label, dim=0)
print(final_tensor.shape)
```
